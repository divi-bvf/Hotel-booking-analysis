```python
#outlier handling -- z score
outliers=[]
def func_outliers(data):
    threshold=3
    mean = np.mean(data)
    std=np.std(data)
    
    for i in data:
        if isinstance(i, (int, float)):       # Check if 'i' is a numeric value
            z_score = (i - mean) / std
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
        
    return outliers
    
```

```python
t=func_outliers(df_new['adr'])
print(t)
```

```python
#outlier handling -- z score
outliers_1=[]
def func1_outliers(data_1):
    thresh=4
    mean = np.mean(data_1)
    std=np.std(data_1)
    
    for i in data_1:
        if isinstance(i, (int, float)):       # Check if 'i' is a numeric value
            z = (i - mean) / std
        z=(i-mean)/std
        if np.abs(z)>thresh:
            outliers_1.append(i)
        
    return outliers_1
    
```

```python
e=func1_outliers(df_new['children'])
print(e)
```

```python
def remove_outliers(data):
    outliers = func_outliers(data)
    # Keep only the values that are not outliers
    cleaned_data = data[~data.isin(outliers)]
    return cleaned_data


# Assuming 'df_new' is a pandas Series or DataFrame containing 'adr' values
# Create a cleaned version without outliers
cleaned_data = remove_outliers(df_new['adr'])
cleaned_data
```

```python
def remove_outliers1(data_1):
    outliers_1 = func1_outliers(data_1)
    # Keep only the values that are not outliers
    cleaned_data = data_1[~data_1.isin(outliers_1)]
    return cleaned_data


# Assuming 'df_new' is a pandas Series or DataFrame containing 'adr' values
# Create a cleaned version without outliers
cleaned_data1 = remove_outliers1(df_new['children'])
cleaned_data1
```

```python
#extracting month from date column using datetime module
df_cleaned['month']=df_cleaned['reservation_status_date'].dt.month
```

```python
#visualizing reservation status monthy
plt.figure(figsize=(16,8))
sx=sns.countplot(x='month',hue='is_canceled',data=df_cleaned,palette='colorblind')
plt.legend(['not canceled','canceled'],bbox_to_anchor=(1,1))
plt.title('Monthly Reservation status')
plt.xlabel('month',color='white')
plt.ylabel('reservation status', color='white')
plt.show()        #august has the lowest cancelation and january has the highest cancelation among them
```

```python
#visualizing the reservation that if they being affected by the increase monthly price 
plt.figure(figsize=(16,8))
sns.barplot(x='month',y='adr',data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR plot')
plt.xlabel('month')
plt.ylabel('ADR')
plt.show()      #january has the highest increase in price according to this plot and also highest cancelation in the above plot.This shows that price has affected the cancelation of reservation
```

```python
#countrywise analysis of most reservation cancaletion
plt.figure(figsize=(10,10))
crty_canceled=df[df['is_canceled']==1]['country'].value_counts()[:10]
plt.title('Top 20 countries with most cancelation')
plt.pie(crty_canceled,autopct='%.2f',labels=crty_canceled.index)
plt.show()
```
